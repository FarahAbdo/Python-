```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import re

options = Options()
options.binary_location = "C:/Program Files/Google/Chrome/Application/chrome.exe"

driver = webdriver.Chrome(options=options, executable_path= 'C:/chromedriver.exe')
driver.get('http://google.com/')      
print("Chrome Browser Invoked")
driver.quit()

# find where chromedriver app is
# s = Service('C:/chromedriver.exe')
# link to our Chrome() class
# driver = webdriver.Chrome(service=s)


def cut(sentence, k):
    
    if len(sentence) >= k:
        cut_list = re.findall(r'.{' + str(k) + '}', sentence)
        rest_chars_num = len(sentence) - 5000*(len(sentence)//5000)
        rest_chars = sentence[len(sentence)-rest_chars_num:]
        cut_list.append(rest_chars)
        return cut_list
    else:
        return  [sentence]



def translate(from_lang_text='English', to_lang_text='Arabic', sentence='How are you!'):
    
    driver.get("https://translate.google.com/")
    time.sleep(4)

    # Set languages

    # Arrow from Lang
    xpath = '//*[@id="yDmH0d"]/c-wiz/div/div[2]/c-wiz/div[2]/c-wiz/div[1]/div[1]/c-wiz/div[1]/c-wiz/div[2]/button/div[3]'
    arrow_from_lang = driver.find_element(By.XPATH, xpath)
    arrow_from_lang.click()
    time.sleep(2)

    # From Lang
    xpath = '//*[@id="yDmH0d"]/c-wiz/div/div[2]/c-wiz/div[2]/c-wiz/div[1]/div[1]/c-wiz/div[2]/c-wiz/div[1]/div/div[2]/input'
    from_lang = driver.find_element(By.XPATH, xpath)
    from_lang.send_keys(from_lang_text)
    from_lang.send_keys(Keys.ENTER)
    time.sleep(2)



    # Arrow to lang
    xpath = '//*[@id="yDmH0d"]/c-wiz/div/div[2]/c-wiz/div[2]/c-wiz/div[1]/div[1]/c-wiz/div[1]/c-wiz/div[5]/button/div[3]'
    arrow_to_lang = driver.find_element(By.XPATH, xpath)
    arrow_to_lang.click()
    time.sleep(2)


    # To Lang
    xpath = '//*[@id="yDmH0d"]/c-wiz/div/div[2]/c-wiz/div[2]/c-wiz/div[1]/div[1]/c-wiz/div[2]/c-wiz/div[2]/div/div[2]/input'
    to_lang = driver.find_element(By.XPATH, xpath)
    to_lang.send_keys(to_lang_text)
    to_lang.send_keys(Keys.ENTER)
    time.sleep(2)





    # Translate a paragraph
    google_translation_ability_num = 5000
    sentence_parts = cut(sentence, google_translation_ability_num)
    output_parts = []
    for part in sentence_parts:
        
        # Write the text to be translated
        xpath = '//*[@id="yDmH0d"]/c-wiz/div/div[2]/c-wiz/div[2]/c-wiz/div[1]/div[2]/div[3]/c-wiz[1]/span/span/div/textarea'
        translate_box = driver.find_element(By.XPATH, xpath)
        translate_box.send_keys(part)
        time.sleep(10)


        # Translated paragraph
        xpath = '//*[@id="yDmH0d"]/c-wiz/div/div[2]/c-wiz/div[2]/c-wiz/div[1]/div[2]/div[3]/c-wiz[2]/div/div[8]/div/div[1]/span[1]'
        translated_text = driver.find_element(By.XPATH, xpath)
        
        
        output = translated_text.text
        output_parts.append(output)
        
        time.sleep(5)
        translate_box.send_keys(Keys.CONTROL + 'a', Keys.BACKSPACE)


        
        
        
        
        
    final_output = ' '.join(output_parts)
    return final_output
```
